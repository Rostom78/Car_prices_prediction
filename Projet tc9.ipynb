```python
#we start with data standarization :
# mileage engine max_power 
#we start with the fuel 
df["fuel"].value_counts()
```

```python
#as we can see we have 3 types of fuel so we use the function below:
df.fuel = df.fuel.map({'Diesel':0, 'Petrol':1,'CNG':2,'LPG':3})
#from sklearn.preprocessing import LabelEncoder
#encoder=LabelEncoder()
#df['fuel']=encoder.fit_transform(df['fuel'])'''
```

```python
df["seller_type"].value_counts()
```

```python
df.seller_type = df.seller_type.map({"Individual":0, "Dealer":1,"Trustmark Dealer":2})
#df['seller_type']=encoder.fit_transform(df['seller_type'])
```

```python
#df['transmission']=encoder.fit_transform(df['transmission'])
```

```python
df["seller_type"].isnull().sum()
```

```python
#we go now to the transmission :
df.transmission = df.transmission.map({"Manual":0, "Automatic":1})
```

```python
df.owner.value_counts()
```

```python
df.owner= df.owner.map({"First Owner":1, "Second Owner":2,"Third Owner":3,"Fourth & Above Owner":4,"Test Drive Car":0})
#df['owner']=encoder.fit_transform(df['owner'])
```

```python
#as we can see it's working
df.owner
```

```python
#Maintenant on enleve les substrings suivantes de engine,mileage et max_power:

#df.engine=df.engine.str.strip(" CC")
#df.engine=df.engine.str.strip(' km/kg')

df.mileage=df.mileage.replace(' km/kg','')
#df.mileage=df.mileage.replace([' km/kg',' kmpl'],'')
df['max_power']=df['max_power'].replace(' bhp','')
```

```python
#on verifie les donnÃ©es nulles qu'on a 
df.isnull().sum()
```

```python
#on remplace les valeurs nulles:
df.seats.fillna(5,inplace=True)
df.seats.isnull().sum()
```

```python
df.engine=df.engine.astype(float)
df.engine.fillna(df.engine.mean(),inplace=True)
df.engine.isnull().sum()
```

```python
df.info()
```

```python
#df.max_power.str.replace('','0')

df.mileage.fillna(df.mileage.mean(),inplace=True)
df.mileage.isnull().sum()
```

```python

df.max_power.fillna(df.max_power.mean(),inplace=True)    
df.max_power.isnull().sum()
```

```python
df.torque.fillna(df.torque.mean(),inplace=True) 
```

```python
df.isnull().sum()
```

```python
# Now we still have to do some feature engineering on our name:
```

```python
#we're gonna extract the 2 first words of our noun as manufacturer and model:
manufacturer = []
modeles = []
for idx,name in zip(range(len(df['name'])),df['name']):
    manufacturer.append(name.split(" ")[0])
    modeles.append(name.split(" ")[1])
df['manufacturer'] = manufacturer
df['modeles'] = modeles
```

```python
#Decision Tree: 
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score

#features extraction
x=df.drop(["selling_price"], axis=1)
y= df["selling_price"]

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=0)
```

```python
from sklearn import metrics

decisiontree = tree.DecisionTreeRegressor(random_state=0)
decisiontree.fit(x_train, y_train) #fitting our model
y_pred=decisiontree.predict(x_test) # evaluating our model
m=metrics.r2_score(y_test,y_pred)
print(m)
dict_score["decisiontree"]=m
print(dict_score)
#Results.append[["dtc",m]]
```

```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor #Importing Random Forest Classifier
from sklearn import metrics # Importing metrics to test accuracy
```

```python
#x=df.drop(["selling_price","mileage"], axis=1)
x=df[['year','transmission','max_power','engine']]
y= df["selling_price"]
#splitting data
 #target
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3) #splitting data with test size of 30%
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
#extract x and y from our data (were going to try )
#trying the linear regression with the more corelated features with our output
x=df["max_power"].values[:,np.newaxis]
y=df["selling_price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=4) #splitting datawith test size of 35%
#applying linear regression on our data
lreg=LinearRegression() #build linear regression model
lreg.fit(x_train,y_train) #fitting the training data
predicted=lreg.predict(x_test) #testing our modelâ€™s performance
print("MSE", mean_squared_error(y_test,predicted))
print("R squared",metrics.r2_score(y_test,predicted))
m3=metrics.r2_score(y_test,predicted)
dict_score["lreg"]=m3
print(dict_score)
```

```python
plt.scatter(x_test.squeeze(),y_test,label="data points")
plt.title("Linear Regression")
plt.ylabel("selling_price")
plt.xlabel("max_power")
plt.plot(x_test,lreg.predict(x_test),color="k")
plt.show()
```

```python
#now we're gonna try the polynomial regression
from sklearn.preprocessing import PolynomialFeatures

y= df["selling_price"].values
x=df.drop(["selling_price"],axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40, random_state=40) #splitting data
lg=LinearRegression()
polynomial=PolynomialFeatures(degree=2)
x_train_fit = polynomial.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = polynomial.fit_transform(x_test)
predicted = lg.predict(x_test_)
print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
m5=metrics.r2_score(y_test,predicted)
dict_score["polynomial"]=m5
print(dict_score)
```
